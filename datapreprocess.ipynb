```python
import pandas as pd
```

```python
df=pd.read_csv("/content/train.csv")
df
```

```python
df.info()
```

```python
df1=df.dropna()
df1.info()
```

```python
df1["Brand"].unique()
```

```python

# One-hot encode multiple columns: 'Brand', 'Material', 'Style', 'Color'
encoded_brands = pd.get_dummies(df1[['Brand', 'Material', 'Style', 'Color']])

# Convert True/False to 1/0 by changing the datatype to int
encoded_brands = encoded_brands.astype(int)

# Drop original columns and concatenate the one-hot encoded columns
df2 = pd.concat([df1.drop(['Brand', 'Material', 'Style', 'Color'], axis=1), encoded_brands], axis=1)

# Print the final encoded DataFrame
print(df2)
```
